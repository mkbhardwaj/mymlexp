One column holds multiple variable 

```python
country_timeseries = pd.read_csv('../data/country_timeseries.csv')
```

```python
country_timeseries.head()
```

```python
# melt down the data 
country_timeseries_long = pd.melt(country_timeseries, 
    id_vars = ['Date', 'Day'],
    var_name='CaseOrDeath_Country',
    value_name='Count')
```

```python
country_timeseries_long.head()
```

```python
countries = country_timeseries_long.CaseOrDeath_Country.str.split('_').str[1]
# one way to do the column spliting is create a contry and 
# and caseordeath series and add it to the dataframe otherwise 
# we can use expand = True in split which will return the dataframe itself 
# which can be further used.
```

```python
splitted_columns = country_timeseries_long.CaseOrDeath_Country.str.split('_', expand=True)
splitted_columns.head()
```

```python
country_timeseries_long[['CasesOrDeaths', 'Country']] = splitted_columns
```

```python
country_timeseries_long.head()
```

Now this data is free from one column holding multiple attribute (country and case/death) but there is one more issue in the data. If we check the column Cases Or Deaths, it is actually holding the value which can be separated as column Cases and Death. So let's resolve Variable are stored in rows not column

```python
# we can pivot the column CasesOrDeath and Count 
country_timeseries_long_pivoted = \
    pd.pivot(country_timeseries_long, 
        index = ['Date', 'Day', 'Country',  'CaseOrDeath_Country'],
        columns= ['CasesOrDeaths'],
        values='Count')
```

```python
country_timeseries_long_pivoted.head()
country_timeseries_long_pivoted =  country_timeseries_long_pivoted.reset_index()
```

```python
country_timeseries_long_pivoted.head()
```

```python
country_timeseries_long_pivoted = country_timeseries_long_pivoted.drop('CaseOrDeath_Country',axis=1)
```

```python
country_timeseries_long_pivoted.head()
```

Billboard data has redundant value for the column year, artist, track, time and date entered for each week. If we are saving this data to some database then we can normalize it and create two separate data frame one holding the track record and other one holdign the ratings.


```python
billboard_track = billboard_long[['year',
    'artist',
    'track',
    'time',
    'date.entered']]
billboard_track = billboard_track.drop_duplicates()
```

```python
billboard_track.head()
```

```python
billboard_track.shape
```

```python
billboard_track['id'] = list(range(len(billboard_track)))
```

```python
billboard_track.head() # we have created the track part 
```

```python
# now create the ratings 
billboard_long_merged = billboard_long.merge(billboard_track,
    on=['year','artist','track', 'time','date.entered'])
```

```python
billboard_long_merged.head()
```

```python
billboard_ratings =  billboard_long_merged[['id', 'week', 'rating']]
```
