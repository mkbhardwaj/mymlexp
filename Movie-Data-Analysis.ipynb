```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline 
```

```python
# load the data 
movieData = pd.read_csv('Movie-Rating-Dataset.csv')
```

```python
# change the datatype of certain columns so we can categorize them later 
movieData.columns = ['Day_Of_Week',\
                     'Director',\
                     'Genre',\
                     'Movie_Title',\
                     'Release_Date',\
                     'Studio',\
                     'Adjusted_Gross',\
                     'Budget',\
                     'Gross',\
                     'IMDB_Rating',\
                     'Movielens_Rating',\
                     'Overseas',\
                     'Overseas_Percentage',\
                     'Profit',\
                     'Profit_Percentage',\
                     'Runtime_In_Minute',\
                     'US',\
                     'Gross_Percentage_US'
                    ]
```

```python
movieData.info()
```

```python
movieData.describe()
```

```python
# filter data sets 
genreToSelect = ['action', 'comedy', 'adventure', 'animation', 'drama']
studioToSelect = ['Buena Vista Studios', 'Sony', 'Universal', 'WB', 'Paramount Pictures', 'Fox']
filteredMovieData = movieData[movieData.Genre.isin(genreToSelect) & movieData.Studio.isin(studioToSelect) ]
```

```python
filteredMovieData.Genre = filteredMovieData.Genre.astype('category')
filteredMovieData.Studio = filteredMovieData.Studio.astype('category')
```

```python
plot = sns.boxplot(data = filteredMovieData, x = 'Genre', y = 'Gross_Percentage_US',  color = 'gray')
swarm = sns.swarmplot(x="Genre", y="Gross_Percentage_US", data=filteredMovieData, hue = 'Studio')
sns.set(rc = {'figure.figsize':(25,10)})
plot.set_xlabel('Genre', fontsize = 18)
plot.set_ylabel('Gross % US', fontsize = 18)
plot.set_title('Domestic Gross % by Genre', fontsize = 25)
plt.legend(loc = 'upper left', bbox_to_anchor=(1,1))
plt.show()
```
